```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing and preparing data

```python
gdl_articles_coverage = pd.read_csv('gdl_world_coverage.csv', parse_dates=['date'])
jdg_articles_coverage = pd.read_csv('jdg_world_coverage.csv', parse_dates=['date'])
```

```python
gdl_articles_coverage = gdl_articles_coverage.set_index('date').sort_index()
jdg_articles_coverage = jdg_articles_coverage.set_index('date').sort_index()
```

```python
gdl_articles_coverage = gdl_articles_coverage.groupby('date').aggregate(np.sum)
jdg_articles_coverage = jdg_articles_coverage.groupby('date').aggregate(np.sum)
```

## Investigation

```python
reduced_gdl_articles_coverage = gdl_articles_coverage.sum(axis=1)
reduced_gdl_articles_coverage = reduced_gdl_articles_coverage.groupby('date').aggregate(np.sum)

reduced_jdg_articles_coverage = jdg_articles_coverage.sum(axis=1)
reduced_jdg_articles_coverage = reduced_jdg_articles_coverage.groupby('date').aggregate(np.sum)
```

```python
plt.plot(reduced_gdl_articles_coverage.as_matrix(), 'x')
plt.plot(reduced_jdg_articles_coverage.as_matrix(), 'x')
plt.show()
```

```python
reduced_gdl_articles_coverage.cumsum().plot()
reduced_jdg_articles_coverage.cumsum().plot()
plt.show()
```

```python
gdl_articles_coverage.loc['1988':'1991'].sum(axis=1).cumsum().plot()
plt.show()
```
